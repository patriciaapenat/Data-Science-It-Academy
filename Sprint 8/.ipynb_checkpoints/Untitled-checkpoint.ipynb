##### Codificació cíclica

La codificació cíclica és una tècnica utilitzada per representar variables categòriques que tenen una naturalesa cíclica, com els mesos de l' any o els dies de la setmana, de manera que es conservi la informació de periodicitat. La idea principal és utilitzar funcions trigonomètriques com el si i el cosinus per crear noves variables numèriques que capturin la periodicitat de la variable original.

###### Data

```python
fb['YRday_published'] = fb['date_published'].dt.dayofyear # Dia de l'any sobre 365-366
fb['month_published'] = fb['date_published'].dt.month
fb['year_published'] = fb['date_published'].dt.year
```

```python
# Obtenim sens i cosins pel month
fb['month_sin'] = np.sin(2 * np.pi * fb['month_published'] / 12)
fb['month_cos'] = np.cos(2 * np.pi * fb['month_published'] / 12)
```

    Ara per obtenir els sens i cosins dels dias per any hem de tenir en compte que les nostres dades van de 2012 a 2018 i tindriem dos anys bisiestos (de traspàs), això ens fa haver de ver la divisió entre 366 enlloc de entre 365

```python
fb['day_sin'] = np.where((fb['year_published'] == 2012) | (fb['year_published'] == 2016),
                         np.sin(2 * np.pi * fb['YRday_published'] / 366),
                         np.sin(2 * np.pi * fb['YRday_published'] / 365))

fb['day_cos'] = np.where((fb['year_published'] == 2012) | (fb['year_published'] == 2016),
                         np.cos(2 * np.pi * fb['YRday_published'] / 366),
                         np.cos(2 * np.pi * fb['YRday_published'] / 365))
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score
from sklearn.metrics.pairwise import euclidean_distances

def evaluar_clustering(model, dades, df_metricas=None, nom_model=None):
    # Realitzar el clustering i obtenir els resultats
    resultats_clustering = model.labels_

    # Obtindre els centroides dels clusters
    centroides = model.cluster_centers_

    # Calcular la distància euclidiana al quadrat entre les dades i els centroides
    distancies_quadrat = np.square(euclidean_distances(dades, centroides))

    # Calcular les mètriques
    coef_silueta = silhouette_score(dades, resultats_clustering)
    indice_davies_bouldin = davies_bouldin_score(dades, resultats_clustering)
    WCSS = np.sum(distancies_quadrat[np.arange(len(dades)), resultats_clustering])
    BCSS = np.sum(distancies_quadrat) - WCSS

    # Crear un DataFrame amb les mètriques
    noves_metricas = pd.DataFrame({
        'Model': [nom_model],
        'Coeficient de Silueta': [coef_silueta],
        'Índex de Davies-Bouldin': [indice_davies_bouldin],
        'WCSS': [WCSS],
        'BCSS': [BCSS]
    })

    # Afegir les noves mètriques al DataFrame existent o crear-ne un de nou si no es proporciona
    if df_metricas is None:
        df_metricas = noves_metricas
    else:
        df_metricas = pd.concat([df_metricas, noves_metricas], ignore_index=True)

    return df_metricas
```

### K Means

K-means és un algoritme de clúster o agrupament que s'utilitza per separar un conjunt de dades en grups o clústers basats en les seves característiques. L'objectiu principal de l'algorisme K-means és agrupar les dades de manera que els punts dins de cada grup siguin similars entre si i els punts en diferents grups siguin diferents.

L'algorisme K-means funciona de la manera següent:

- Seleccioneu un nombre predefinit de grups (k) que voleu obtenir a l'agrupació.
- Inicialitza k centroides de manera aleatòria a l'espai de característiques.
- Assigna cada punt de dades al centreide més proper en funció de la distància.
- Recalcula els centreides de cada grup com la mitjana dels punts assignats a aquest grup.
- Repetiu els passos 3 i 4 fins que els centreides dels grups convergeixin o s'arribi a un nombre màxim d'iteracions.
- El resultat final de l'algorisme K-means és una partició de les dades en grups k, on cada punt de dades està assignat a un dels grups. Els centroides dels grups representen les posicions centrals o mitges dels punts assignats a cada grup.

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, n_init=10)
kmeans.fit(ex1_pca)
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_
```

Expliquem breument les metriques

- Coeficient de Silueta: El coeficient de silueta és una mesura de la qualitat de l'agrupació. Un valor més proper a 1 indica una bona separació i cohesió dels grups, cosa que significa que els punts dins de cada grup estan més a prop entre si que dels punts en altres grups. En aquest cas, el coeficient de silueta és de 0.869317, cosa que indica que l'agrupació té una bona separació i cohesió.

- Índex de Davies-Bouldin: L'índex de Davies-Bouldin és també una mesura de la qualitat de l'agrupació. Un valor més baix indica una millor qualitat de l'agrupació, on els grups estan més separats i els punts dins de cada grup són més similars entre si. En aquest cas, l'índex de Davies-Bouldin és de 0,78222, cosa que indica una bona qualitat de l'agrupació.

- WCSS: La suma de quadrats dins del cluster (WCSS) és una mesura de la dispersió dels punts dins de cada grup. Com més baix sigui el valor de WCSS, més compactes i similars seran els punts dins de cada grup. En aquest cas, el valor de WCSS és de 6.523661e+06.

- BCSS: La suma de quadrats entre clusters (BCSS) és una mesura de la separació entre els grups. Com més gran sigui el valor de BCSS en comparació amb WCSS, més separats estaran els grups. En aquest cas, el valor de BCSS és de 9.314246e+08.

Els resultats indiquen que l'agrupació té una bona qualitat, amb grups ben separats i cohesius. Els valors de les mètriques mostren que els punts dins de cada grup estan a prop entre si i els grups estan ben separats entre si.

## Exercici 2
Classifica els diferents registres utilitzant l'algorisme de clustering jeràrquic.

```python
import pandas as pd
import numpy as np
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
from sklearn.metrics.pairwise import euclidean_distances

def evaluar_clustering(model, datos, metricas_df=None, nombre_modelo=None):
    try:
        # Realizar el clustering y obtener los resultados
        resultados_clustering = model.labels_

        # Obtener los centroides de los clusters
        centroides = model.cluster_centers_

        # Calcular la distancia euclidiana al cuadrado entre los datos y los centroides
        distancias_cuadrado = np.square(euclidean_distances(datos, centroides))

        # Calcular las métricas
        try:
            coef_silueta = silhouette_score(datos, resultados_clustering)
        except:
            coef_silueta = np.nan
        try:
            indice_davies_bouldin = davies_bouldin_score(datos, resultados_clustering)
        except:
            indice_davies_bouldin = np.nan
        try:
            indice_calinski_harabasz = calinski_harabasz_score(datos, resultados_clustering)
        except:
            indice_calinski_harabasz = np.nan

        WCSS = np.sum(distancias_cuadrado[np.arange(len(datos)), resultados_clustering])
        BCSS = np.sum(distancias_cuadrado) - WCSS

        # Crear un DataFrame con las métricas
        nuevas_metricas = pd.DataFrame({
            'Modelo': [nombre_modelo],
            'Coeficiente de Silueta': [coef_silueta],
            'Índice de Davies-Bouldin': [indice_davies_bouldin],
            'Índice Calinski-Harabasz': [indice_calinski_harabasz],
            'WCSS': [WCSS],
            'BCSS': [BCSS]
        })

        # Agregar las nuevas métricas al DataFrame existente o crear uno nuevo si no se proporciona
        if metricas_df is None:
            metricas_df = nuevas_metricas
        else:
            metricas_df = pd.concat([metricas_df, nuevas_metricas], ignore_index=True)

    except AttributeError:
        print("Error")

    return metricas_df
```
