#### Col·linealitat

El Factor d'Inflació de la Variança (Variance Inflation Factor, VIF) és una mesura utilitzada per detectar la multicol·linealitat en un model de regressió. El VIF avalua quant s'infla la variància d'un coeficient de regressió a causa de la multicol·linealitat.

El VIF es calcula per a cada variable predictora al model. Per calcular el VIF d'una variable en particular, es fa una regressió lineal utilitzant totes les altres variables predictores com a predictors. Tot seguit, s'obté el coeficient de determinació (R^2) d'aquesta regressió. El VIF es calcula com l'invers de (1 - R^2).

Un VIF igual a 1 indica que no hi ha multicol·linealitat entre la variable i les altres variables predictores. Com més gran sigui el VIF, més gran serà la multicol·linealitat. Es considera que un VIF superior a 5 o 10 indica la presència de multicol·linealitat significativa.

```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Agrega una columna constante (intercepto) a x_train
x_train_with_const = sm.add_constant(x_train)

print("VIF para x['RAD']:", variance_inflation_factor(x_train_with_const.values, x_train_with_const.columns.get_loc('RAD')))
print("VIF para x['TAX']:", variance_inflation_factor(x_train_with_const.values, x_train_with_const.columns.get_loc('TAX')))
print("VIF para x['CHAS']:", variance_inflation_factor(x_train_with_const.values, x_train_with_const.columns.get_loc('CHAS')))
```

Mirant els valors del VIF podem concloure que existeix multicol·linealitat significativa a RAD y TAX, tornem als gràfics de disperssió per saber si la relació es complexa

### Regressió lineal multiple

```python
import statsmodels.api as sm

# Crear un diccionari que ens ajudarà a poder recuperar cada linmodel més endavant
linmodels = {}

# Iterar sobre columnes específiques
for column in x_train[['CRIM', 'ZN', 'INDUS', 'NOX', 'RM', 'AGE', 'DIS', 'PRATIO', 'B_1000', 'LSTAT']]:
    # Ajustar el model
    linmodel = sm.OLS(y_train, x_train[column]).fit()

    # Almacenar el modelo en el diccionario usando el nombre de la columna como clave
    linmodels[column] = linmodel

# Accedir als models
for column, model in linmodels.items():
    print(f"Variable: {column}")
    print(model.summary())
    print()
    print()
```

### Regressió polinòmica
Les regressions polinòmiques són una extensió de la regressió lineal en la qual es permet modelar relacions no lineals entre les variables predictores (independents) i la variable objectiu (dependent). En lloc d' ajustar una línia recta com en la regressió lineal, en la regressió polinòmica s' utilitzen polinomis de grau superior per capturar patrons més complexos en les dades.

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
# Creació d'objecte polynomialfeatures con un grado polinomial de 2, també es fa la transformació en una matriu polinòmica
x_train_poly = PolynomialFeatures(degree=2).fit_transform(x_train['CHAS'].values.reshape(-1, 1))
# Ajustem el model
lin_poly = LinearRegression().fit(x_train_poly, y_train)
```

```python
# Fem un gràfic
x_range = np.linspace(min(x_train['CHAS']), max(x_train['CHAS']), 100).reshape(-1, 1)
x_range_poly = PolynomialFeatures(degree=2).fit_transform(x_range)
y_predpol = lin_poly.predict(x_range_poly)

plt.scatter(x_train['CHAS'], y_train)
plt.plot(x_range, y_predpol, color='red')
plt.xlabel('CHAS')
plt.ylabel('MEDV')
plt.title('Regresión polinómica')
plt.show()
```

### Regressió de random forest

Les regressions d'arbre de decisió són un tipus de model de regressió que utilitza l'estructura d'un arbre de decisió per fer prediccions. En lloc d'ajustar una línia recta com en la regressió lineal, els arbres de decisió divideixen l'espai de característiques en regions més petites i fan prediccions basades en les característiques que cauen a cada regió.

Un arbre de decisió consta de nodes i branques. Cada node representa una característica o condició sobre una característica, i cada branca representa el resultat d'aquesta condició. L'arbre es construeix dividint recursivament les dades en funció de les característiques més informatives a cada pas, de manera que es minimitzi l'error de predicció.

Quan s'aplica una regressió d'arbre de decisió, una predicció numèrica s'assigna a cada full de l'arbre. Durant la predicció, se segueix el camí des de l'arrel fins a un full en funció dels valors de les característiques, i s'utilitza el valor assignat a aquest full com a predicció.

Els arbres de decisió són models versàtils i poden manejar tant característiques categòriques com numèriques. També són capaços de capturar relacions no lineals a les dades. No obstant això, poden ser propensos al sobreajustament si no es controla adequadament, cosa que pot portar a un rendiment deficient en conjunts de dades noves.

    El concepte de línia base (baseline) s'utilitza a l'aprenentatge automàtic com a punt de referència per avaluar el rendiment d'un model predictiu. Al codi que vau proporcionar, es calcula l'error de línia base utilitzant la mitjana històrica com a prediccions base.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
```

```python
# Crear una instancia del modelo de Random Forest per RAD
rf_modelrad = RandomForestRegressor(random_state=42)

# Entrenar el modelo solo con la columna 'rad' de los datos de entrenamiento
rf_modelrad.fit(x_train['RAD'].values.reshape(-1, 1), y_train)

# Realizar predicciones solo con la columna 'rad' de los datos de prueba
rf_predictionsrad = rf_modelrad.predict(x_test['RAD'].values.reshape(-1, 1))

# Calcular el error de línea base (usando el promedio histórico) en el conjunto de prueba
baseline_errorrad = np.mean(abs(np.full_like(y_test, y_train.mean()) - y_test))
```

```python
# Graficar els valors reals
plt.scatter(range(len(y_test)), y_test, color='#2C3172', label='Valores reales')

# Graficar les prediccions del model de Random Forest
plt.scatter(range(len(y_test)), rf_predictionsrad,color='#A5CD90', label='Prediccions')

plt.xlabel('Instàncies')
plt.ylabel('Variable objectiu')
plt.title('Comparació entre valoes reals i prediccions RAD')
plt.legend()
plt.show()
```

```python
# Crear una instancia del modelo de Random Forest per TAX
rf_modeltax = RandomForestRegressor(random_state=42)

# Entrenar el modelo solo con la columna 'rad' de los datos de entrenamiento
rf_modeltax.fit(x_train['TAX'].values.reshape(-1, 1), y_train)

# Realizar predicciones solo con la columna 'rad' de los datos de prueba
rf_predictionstax = rf_modeltax.predict(x_test['TAX'].values.reshape(-1, 1))

# Calcular el error de línea base (usando el promedio histórico) en el conjunto de prueba
baseline_errortax = np.mean(abs(np.full_like(y_test, y_train.mean()) - y_test))
```

```python
# Graficar los valores reales
plt.scatter(range(len(y_test)), y_test, color='#2C3172', label='Valores reales')

# Graficar las predicciones del modelo de Random Forest
plt.scatter(range(len(y_test)), rf_predictionstax, color='#A5CD90', label='Predicciones')

plt.xlabel('Instancias')
plt.ylabel('Variable objetivo')
plt.title('Comparación entre valores reales y predicciones TAX')
plt.legend()
plt.show()
```

### Comparativa

#### MSE

El MSE (Mean Squared Error, en inglés) es una métrica utilizada para evaluar la calidad de un modelo de regresión. Mide el promedio de los errores cuadráticos entre los valores reales observados (y) y los valores predichos por el modelo (\hat{y}).

El MSE se utiliza comúnmente en problemas de regresión para comparar diferentes modelos o evaluar la calidad de las predicciones del modelo. Cuanto menor sea el valor del MSE, mejor será la calidad del modelo, ya que indica que los valores predichos se acercan más a los valores reales.

```python
from collections import OrderedDict
```

```python
mse_results = OrderedDict(sorted(mse_results.items(), key=lambda x: x[1]))
```

```python
mse_results
```

```python
# Obtener el MSE mínim y màxim
min_mse = min(mse_results.values())
max_mse = max(mse_results.values())

# Imprimir els resultats
print("MSE mínimo:", min_mse)
print("MSE máximo:", max_mse)
print("MSE promedio:", sum(mse_results.values()) / len(mse_results))
print("Modelo con MSE mínimo:", next(key for key, value in mse_results.items() if value == min_mse))
print("Modelo con MSE máximo:", next(key for key, value in mse_results.items() if value == max_mse))
```

#### R^2

El coeficient de determinació, comunament conegut com a R^2, és una mesura estadística que indica la proporció de la variància de la variable dependent que és predictible a partir de les variables independents en un model de regressió. En altres paraules, R^2 representa què tan bé s'ajusten els valors observats als valors predits pel model de regressió.

El valor de R^2 es troba en el rang de 0 a 1. Un valor de 0 indica que el model no explica cap variabilitat a la variable dependent, mentre que un valor de 1 indica que el model explica tota la variabilitat a la variable dependent.

```python
r2_results = OrderedDict(sorted(r2_results.items(), key=lambda x: x[1]))
```

```python
r2_results
```

```python
# Imprimir los resultados
print("Modelo con R^2 más alto:", next(iter(r2_results)))
print("R^2 más alto:", r2_results[next(iter(r2_results))])
print("Modelo con R^2 más bajo:", next(reversed(r2_results)))
print("R^2 más bajo:", r2_results[next(reversed(r2_results))])
```
